## Exploratory Data Analysis (EDA)

Change the year, month, date and time columns into oen date time column

```python
def process_dataframe(df):
  df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
  df.set_index('datetime', inplace=True)
  redundant_cols = ['year', 'month', 'day', 'hour']
  df.drop(redundant_cols, axis=1, inplace=True)
  return df
```

```python
df = process_dataframe(df)
print(df)
```

### Explore ...

Now we can count the pollution score by using the standardized data

```python
def calculate_pollution_score(data, weights=None):
  pollutant_columns = ["PM2.5", "PM10", "SO2", "NO2", "CO", "O3"]
  pollutant_data = data[pollutant_columns]

  standardized_data = (pollutant_data - pollutant_data.mean()) / pollutant_data.std()

  if weights:
    standardized_data = standardized_data * weights

  pollution_score = standardized_data.sum(axis=1)
  data["pollution_score"] = pollution_score

  return data
```

```python
df = calculate_pollution_score(df.copy())
print(df)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
df['pollution_score'].resample('D').mean().plot(title='Daily Pollution Score Trend')
plt.show()

df['pollution_score'].resample('M').mean().plot(title='Monthly Pollution Score Trend')
plt.show()

df['pollution_score'].resample('Y').mean().plot(title='Yearly Pollution Score Trend')
plt.show()
```

### Pertanyaan 2:

```python
corr_matrix = df[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

## Conclusion

- There is a trend that occurs in CO pollutants every year. The CO pollutant content increases every year, although in 2016 the CO pollutant level dropped. But overall, the trend is upward.
- From the corelation heatmap above, we can see that there is a strong relationship between PM2.5 and PM10, C0 and PM2.5 pollutants. In addition, we can also see that there is a strong relationship between TEMP and DEWP. PRES and TEMP and DEWP and PRES also have a strong relationship, except that the relationship is inversely proportional.
